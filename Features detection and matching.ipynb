```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
I = cv2.imread("feature1.jpg", cv2.IMREAD_COLOR)
Igray = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)
sift = cv2.SIFT_create(100)
orb = cv2.ORB_create(100)
Ifp_sift = sift.detect(Igray)
Ifp_orb = orb.detect(Igray)
Iout_sift = cv2.drawKeypoints(I, Ifp_sift, None, color=(0, 255, 0), flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
Iout_orb = cv2.drawKeypoints(I, Ifp_orb, None, color=(0, 255, 0), flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
cv2_imshow(Iout_sift)
cv2_imshow(Iout_orb)
```

```python
# Load images
I1 = cv2.imread("match2-2.jpg", cv2.IMREAD_COLOR)
I2 = cv2.imread("match2-1.jpg", cv2.IMREAD_COLOR)
I1gray = cv2.cvtColor(I1, cv2.COLOR_BGR2GRAY)
I2gray = cv2.cvtColor(I2, cv2.COLOR_BGR2GRAY)

# Extract features
orb = cv2.ORB_create(nfeatures=2000)
I1fp, I1des = orb.detectAndCompute(I1gray, None)
I2fp, I2des = orb.detectAndCompute(I2gray, None)

# Match
matcher = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck = False)

# Find KNN matches with k = 2
matches = matcher.knnMatch(I1des, I2des, k = 2)
# Select good matches
knn_ratio = 0.7
good = []
for m in matches :
  if len(m) > 1:
    if m[0].distance < knn_ratio * m[1].distance :
      good.append(m[0])
matches = good

# Draw matches
num_matches = 10
matches = sorted(matches, key = lambda x: x.distance )
Imatch = cv2.drawMatches(I1, I1fp, I2, I2fp, matches[: num_matches], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS, matchColor = (0 ,255 ,0))
plt.figure(figsize=(16, 6), dpi=100)
plt.grid(False)
plt.imshow(Imatch)

MIN_MATCH_COUNT = 10
if len (matches) < MIN_MATCH_COUNT :
  print("Not enough matches.")
# Create arrays of point coordinates
I1pts = np.float32([I1fp[m.queryIdx].pt 
  for m in matches]).reshape(-1, 1, 2)
I2pts = np.float32([I2fp[m.trainIdx].pt
  for m in matches]).reshape(-1, 1, 2)

# Run RANSAC method
M, mask = cv2.findHomography(I1pts, I2pts, cv2.RANSAC , 5)
mask = mask.ravel().tolist()

# Image corners
h, w = I2.shape [:2]
I1box = np.float32([[0, 0], [0, h - 1],
  [w - 1, h - 1], [w - 1 , 0]]). \
  reshape(-1, 1, 2)
I1to2box = cv2.perspectiveTransform (I1box, M)
# Draw a red box on the second image
I2res = cv2.polylines(I2, [np.int32 (I1to2box)], True, (0, 0, 255), 5, cv2.LINE_AA)
cv2_imshow(I2res)

Itrans = cv2.drawMatches(I1, I1fp, I2res, I2fp, matches, None, matchesMask=mask, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS, matchColor = (0 ,255 ,0))
cv2_imshow(Itrans)
```
