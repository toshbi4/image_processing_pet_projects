```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
I = cv2.imread("faces2.jpg", cv2.IMREAD_COLOR)
Igray = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)
detector = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
faces = detector.detectMultiScale(Igray, scaleFactor = 1.27, minNeighbors = 3)
Iout1 = I.copy()
for (x, y, w, h) in faces:
  Iout1 = cv2.rectangle(Iout1, (x, y, w, h), (0, 255, 255), thickness=2)
cv2_imshow(Iout1)
print(len(faces))

Iout2 = I.copy()
# Load eyes cascade
eye_detector = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_eye.xml")
mouth_detector = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_smile.xml")
# For each face use it as a ROI
# and detect eyes
for(x, y, w, h ) in faces:
  Iface = I [ y : y + h , x : x + w ]
  eyes = eye_detector.detectMultiScale(Iface, scaleFactor = 1.05)
  mouthes = mouth_detector.detectMultiScale(Iface, scaleFactor = 2.9)
  for (x2, y2, w2, h2) in eyes:
    Iout2 = cv2.rectangle(Iout2, (x + x2, y + y2, w2, h2), color = (255, 0, 0), thickness=2)
  for (x2, y2, w2, h2) in mouthes:
    Iout2 = cv2.rectangle(Iout2, (x + x2, y + y2, w2, h2), color = (0, 0, 255), thickness=2)
cv2_imshow(Iout2)

Iout3 = I.copy()
for(x, y, w, h ) in faces:
  Iface_top = Igray [ y : y + h * 2 // 3 , x : x + w ]
  Iface_bottom = Igray [ y + h * 1 // 2 : y + h, x : x + w ]
  cv2_imshow(Iface_bottom)
  eyes = eye_detector.detectMultiScale(Iface_top, scaleFactor = 1.05)
  mouthes = mouth_detector.detectMultiScale(Iface_bottom, scaleFactor = 2.35)
  print(y)
  for (x2, y2, w2, h2) in eyes:
    Iout3 = cv2.rectangle(Iout3, (x + x2, y + y2, w2, h2), color=(255, 0, 0), thickness=2)
  for (x2, y2, w2, h2) in mouthes:
    Iout3 = cv2.rectangle(Iout3, (x + x2, y + y2 + h * 1 // 2, w2, h2), color = (0, 0, 255), thickness=2)
cv2_imshow(Iout3)
```
